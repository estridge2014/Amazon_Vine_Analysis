```python
#test save through colab
#Read in the vine_table.csv file as a DataFrame
path = "/Users/savannahestridge/Desktop/class/Amazon_Vine_Analysis/vine_table.csv"
df = pd.read_csv(path)
df
```

```python
#1. Filter the data and create a new DataFrame or table to retrieve all
#the rows where the total_votes count is equal to or greater than 20 to pick reviews
#that are more likely to be helpful and to avoid having division by zero errors later on.

helpful_vote_df = df[df["total_votes"] >= 20]
helpful_vote_df
```

```python
#2.Filter the new DataFrame or table created in Step 1 and create a new DataFrame
#or table to retrieve all the rows where the number of helpful_votes divided by
#total_votes is equal to or greater than 50%.

hv_df = helpful_vote_df[(helpful_vote_df['helpful_votes']/helpful_vote_df['total_votes']) >= 0.5]
hv_df
```

```python
#3.Filter the DataFrame or table created in Step 2, and create a new DataFrame or table
#that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

vine_y_df = hv_df[(hv_df['vine'] == 'Y')]
vine_y_df
```
